# Using MOOSE on [binder](https://mybinder.org/)

[binder](https://mybinder.org/) is a free online service for running jupyter notebooks. Every notebook in this repository has a binder button on top like this:  [![Binder](./binder_logo.png)](https://mybinder.org/v2/gh/MooseNeuro/moose-notebooks/HEAD?labpath=MOOSE_on_Binder.ipynb).

This will take you to the binder website and start a session with that specific notebook. This repository is configured so that binder will automatically install the requirements including `pymoose`  in your environment. Once the environment is ready, you can start running the code.

If you want to run multiple notebooks, it is simpler to open the entire repository on binder by either

a) insert the relative address of this repository  (`MooseNeuro/moose-notebooks`) in the "GitHub repository name or URL" field in the binder landing page and click the "Launch" button, or

b) enter the URL `https://mybinder.org/v2/gh/MooseNeuro/moose-notebooks/HEAD` in your browser's address bar, or

c) click this link [![Binder](https://mybinder.org/badge_logo.svg)](https://mybinder.org/v2/gh/MooseNeuro/moose-notebooks/HEAD).

This will open a tab with the JupyterLab user interface with all the requirements installed. Like your desktop version of JupyterLab, you will see a list of files and folders on the left pane, and you can double click any file to open it in the editor pane. If it is a Jupyter notebook file (filename extension `.ipynb`), you can run the code cells in the usual manner.
