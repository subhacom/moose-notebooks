# Stability and chemical reactions

If you gently drop a ball in a bowl, it will roll around a bit and then settle at the bottom. If you nudge it a bit, it will roll back to that position. If we consider the position of the ball to represent its state, then the bottom of the bowl is its steady state. 

*Image here*

Like the ball-bowl system, we can think of reagents undergoing chemical reaction in a beaker as a chemical system. The concentrations if the reagents can represent the state of the system. Let us consider a simple chemical reaction ([Ferrel and Wen, 2001](https://doi.org/10.1063/1.1349894)):

$A \xrightleftharpoons[I]{S} B$

where $S$ catalyzes the conversion of $A$ into $B$, and $I$ catalyzes the reverse reaction.

The rate of production of $B$ would be $\frac{dB}{dt} = k_{f} \times [A]$
